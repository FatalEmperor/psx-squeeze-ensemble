# Parameter Sensitivity Analysis

How win rate and CAGR respond to changes in:
- `sqz_len` (Squeeze period)
- `adx_threshold` (ADX cutoff)
- `min_sqz_bars` (Minimum squeeze duration)
- `atr_mult` / `tp_mult` (Risk-reward ratio)

```python
import sys, itertools
import pandas as pd
sys.path.insert(0, '../src')

from data.fetcher import fetch_data
from squeeze.indicators import compute_indicators
from squeeze.filters import apply_rule_filters
from squeeze.signals import generate_base_signal, generate_rule_filtered_signal
from backtest.engine import backtest

SYMBOL = 'PSO'
df_raw = fetch_data(SYMBOL)
```

```python
# Sweep ADX threshold
results = []
for adx_thresh in [15, 20, 25, 30, 35]:
    df = compute_indicators(df_raw)
    df = apply_rule_filters(df, adx_threshold=adx_thresh)
    df = generate_base_signal(df)
    df = generate_rule_filtered_signal(df)
    r  = backtest(df, 'sig_rules', SYMBOL)
    results.append({'adx_threshold': adx_thresh, 'trades': r['total_trades'],
                    'win_rate': r['win_rate'], 'cagr': r['cagr_pct'],
                    'max_dd': r['max_drawdown_pct']})

pd.DataFrame(results)
```
